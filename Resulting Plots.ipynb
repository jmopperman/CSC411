### Calculate the coefficient errors

```python
ct0 = 0
ct = 0
error=[]
for i in range(len(Results)-1):
    if ct == 10:
        break
    ## PARAMETERIZE ##
    # error = abs(run0_a1 - runi_a1) + abs(run0_a2 - runi_a2 + abs(run0_b1 - runi_b1) + abs(run0_b2 - runi_b2)
    error.append(abs(np.array(Results.loc['run_'+str(ct0), 'a'+str(1)]) - np.array(Results.loc['run_'+str(ct), 'a'+str(1)]))
                + abs(np.array(Results.loc['run_'+str(ct0), 'a'+str(2)]) - np.array(Results.loc['run_'+str(ct), 'a'+str(2)])) 
                 + abs(np.array(Results.loc['run_'+str(ct0), 'b'+str(1)]) - np.array(Results.loc['run_'+str(ct), 'b'+str(1)])) 
                 + abs(np.array(Results.loc['run_'+str(ct0), 'b'+str(2)]) - np.array(Results.loc['run_'+str(ct), 'b'+str(2)])))
    ct +=1
```

```python
fig = plt.figure(figsize=(8,10))
ax1 = fig.add_subplot(2,1,1)
ax1.set_ylabel("Coefficient Error")
ax1.set_title("TCL_ARX_Differential_Evolution_Step_Input")
ax2 = fig.add_subplot(2,1,2)
ax2.set_ylabel("Runtime (s)")
ax2.set_xlabel("Standard Deviation (sigma)")
ax1.scatter(np.array(Results.std_dev[1:]), np.concatenate(error, axis=0 )[1:])
ax2.scatter(Results.std_dev, Results.Runtime0)
```
