```python
import matplotlib.pyplot as plt
import pandas
import numpy as np
```

```python
Data = pandas.read_excel('C:/Users/annam/Desktop/Johanna/CSC411/Comparison-of-Identification-Techniques/Results/TCL_ARX_order2_DIFEV.xlsx', sheet_name="Data")
Results = pandas.read_excel('C:/Users/annam/Desktop/Johanna/CSC411/Comparison-of-Identification-Techniques/Results/TCL_ARX_order2_DIFEV.xlsx', sheet_name="Results")
```

```python
ct0 = 0
ct = 0
error = []
for i in range(len(Results)-1):
    if ct == 10:
        break
    error.append(abs(np.array(Results.loc['run_'+str(ct0), 'a'+str(1)]) - np.array(Results.loc['run_'+str(ct), 'a'+str(1)])))
    ct +=1
```

```python
fig = plt.figure(figsize=(8,10))
ax1 = fig.add_subplot(2,1,1)
ax1.set_ylabel("Coefficient Error")
ax1.set_title("TCL_ARX_Differential_Evolution_Step_Input")
ax2 = fig.add_subplot(2,1,2)
ax2.set_ylabel("Runtime")
ax2.set_xlabel("Standard Deviation (sigma)")
ax1.scatter(np.array(Results.std_dev[1:]), np.concatenate(error, axis=0 )[1:])
ax2.scatter(Results.std_dev, Results.Runtime0)
```
